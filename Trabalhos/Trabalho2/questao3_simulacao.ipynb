```python
import numpy as np


A = 0
B = 1
N = 100000

count_aceitos = 0
count_total = 0

amostras = []

# ACEITACAO-REJEICAO
while len(amostras) < N:

    # gera Y ~ U(0,1)
    y = np.random.uniform(0, 1)

    # gera U ~ U(0,1)
    u = np.random.uniform(0, 1)

    count_total += 1

    # regra: aceita se U <= y^2
    if u <= y**2:
        amostras.append(y)
        count_aceitos += 1

amostras = np.array(amostras)


# TAXA DE ACEITACAO
taxa = count_aceitos / count_total
print("taxa de aceitacao (simulada) =", taxa)

# valor teorico
taxa_teorica = 1/3
print("taxa de aceitacao (teorica) =", taxa_teorica)


# MEDIA (teste)

media_simulada = np.mean(amostras)
print("media simulada =", media_simulada)

# valor teorico E[X] = 3/4
print("media teorica =", 0.75)
```
