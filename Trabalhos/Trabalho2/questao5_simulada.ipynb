```python
import numpy as np

N = 100000  # número de simulações
sucesso = 0

for i in range(N):
    d1 = np.random.randint(1, 7)  # dado 1
    d2 = np.random.randint(1, 7)  # dado 2

    soma = d1 + d2

    if soma > 9:
        sucesso += 1

prob_simulada = sucesso / N


#Essa parte está no PDF também
# valor teórico
# soma > 9 -> 10, 11, 12
# combinações:
# 10 -> (4,6), (5,5), (6,4) = 3
# 11 -> (5,6), (6,5) = 2
# 12 -> (6,6) = 1
# total = 6 casos favoráveis de 36 possíveis

prob_teorica = 6 / 36

print("Probabilidade simulada =", prob_simulada)
print("Probabilidade teórica =", prob_teorica)
print("Erro absoluto =", abs(prob_simulada - prob_teorica))
```
