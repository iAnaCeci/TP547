```python
import numpy as np
import matplotlib.pyplot as plt

N_n = [10**2, 10**3, 10**4, 10**5]
integral_real = 2*np.log(2) - 1

integral = []
erro = []

for N in N_n:
    x = np.random.uniform(0, 1, N)
    integral_calculada=sum(np.log(1+x))/N

    integral.append(integral_calculada)
    erro.append(abs(integral_real - integral_calculada)) #apenas valor absoluto (magnitude) de erro, poderia ser positivo ou negativo

plt.plot(N_n, erro, 'o-')
plt.xlabel('Número de amostras (N)')
plt.ylabel('Erro absoluto')
plt.title('Erro vs N')
plt.xscale('log') #escala logaritmica para melhor visualização
plt.yscale('log') #escala logaritmica para melhor visualização
plt.grid(True)
plt.show()
```
