```python
import numpy as np
import matplotlib.pyplot as plt

#Monte Carlo

N=10

x = np.random.uniform(0, 1, N)
integral_calculada=sum((1+x)**(-2))/N

print('Integral via Monte Carlo: {}'.format(integral_calculada))
```

```python
# Importance Sampling para p(x) = A*(1+x)^-1

# Gerar u uniforme em [0,1]
u = np.random.uniform(0, 1, N)

# Inversão da CDF: x = exp(u*ln2) - 1
x = np.exp(u * np.log(2)) - 1

# Calculando f(x)/p(x)
estimado = (np.log(2)) / (1 + x)

# 4️⃣ Estimativa da integral
integral_estimada = np.mean(estimado)

print("Integral via Importance Sampling: {}".format(integral_estimada))
```

```python
#Erro Monte carlo
print('Erro Monte Carlo: {}'.format(abs(integral_calculada - 0.5)))

#Erro Importance Sampling
print('Erro Importance Sampling: {}'.format(abs(integral_estimada - 0.5)))
```
