```python
import numpy as np

N = 100000  # número de simulações

sucesso_sem = 0
sucesso_com = 0

cartas = ['A']*12 + ['B']*12 + ['C']*12 + ['D']*12

# =========================
# SEM REPOSIÇÃO
# =========================
for i in range(N):
    retirada = np.random.choice(cartas, 7, replace=False)

    if ('B' in retirada) and ('D' in retirada):
        sucesso_sem += 1

prob_sem = sucesso_sem / N

# =========================
# COM REPOSIÇÃO
# =========================
for i in range(N):
    retirada = np.random.choice(cartas, 7, replace=True)

    if ('B' in retirada) and ('D' in retirada):
        sucesso_com += 1

prob_com = sucesso_com / N

print("Probabilidade sem reposição =", prob_sem)
print("Probabilidade com reposição =", prob_com)
```
