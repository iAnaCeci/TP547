```python
import numpy as np

lambda1 = 4      # média de falhas em 1 hora
lambda2 = 8      # média de falhas em 2 horas
N = 10000        # número de amostras

av1 = np.array([])   # amostras para 1 hora
av2 = np.array([])   # amostras para 2 horas

# 1 HORA
x = np.random.uniform(0, 1, N)

for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F = pr

    while ix >= F:
        pr = lambda1 / (i + 1) * pr
        F = F + pr
        i = i + 1

    av1 = np.append(av1, i)


# 2 HORAS

y = np.random.uniform(0, 1, N)

for iy in y:
    i = 0
    pr = np.exp(-lambda2)
    F = pr

    while iy >= F:
        pr = lambda2 / (i + 1) * pr
        F = F + pr
        i = i + 1

    av2 = np.append(av2, i)



# a) exatamente 6 falhas em 1 hora
prob_a = np.sum(av1 == 6) / N

# b) no máximo 2 falhas em 1 hora
prob_b = np.sum(av1 <= 2) / N

# c) mais de 5 falhas em 2 horas
prob_c = np.sum(av2 > 5) / N

# d) valor esperado e variância
esperado_1h = np.mean(av1)
variancia_1h = np.var(av1)

print("Resultados da simulação:")
print("a) P(X = 6) =", prob_a)
print("b) P(X <= 2) =", prob_b)
print("c) P(X > 5) em 2 horas =", prob_c)
print("d) Valor esperado =", esperado_1h)
print("   Variância =", variancia_1h)
```
