```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000

# gerar U uniforme
u = np.random.rand(n)

# transformação inversa
x = np.sqrt(u)

# histograma normalizado
plt.hist(x, bins=30, density=True)

# pdf teórica
X = np.linspace(0,1,1000)
F = 2*X

plt.plot(X, F)

plt.title("Método da Inversa")
plt.xlabel("X")
plt.ylabel("F(X)")

plt.show()
```
