```python
import numpy as np
import matplotlib.pyplot as plt

x = 5
x1 = np.array([x])
n = 10
a = 11
c = 7
m = 32

for i in range(n):
    x = (a * x + c) % m
    x1 = np.append(x1, x)

print("Sequência gerada:")
print(x1)

ind = np.arange(n + 1)
plt.bar(ind, x1)
plt.xlabel("Índice n")
plt.ylabel("Xn")
plt.title("Gerador Linear Congruente")
plt.show()
```
