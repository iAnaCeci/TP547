```python
import numpy as np

p = 0.08   # probabilidade de erro
n = 20     # número de pacotes
N = 10000  # número de amostras

c = p / (1 - p)
av = np.array([])

x = np.random.uniform(0, 1, N)

for ix in x:
    i = 0
    pr = pow((1 - p), n)   # P(X=0)
    F = pr

    while ix >= F:
        pr = (c * (n - i) / (i + 1)) * pr
        F = F + pr
        i = i + 1

    av = np.append(av, i)

# a) exatamente 3 erros
prob_a = np.sum(av == 3) / N

# b) no máximo 2 erros
prob_b = np.sum(av <= 2) / N

# c) mais de 5 erros
prob_c = np.sum(av > 5) / N

# d) média e variância
media = np.mean(av)
variancia = np.var(av)

print("Amostras geradas:")
print(av)

print("\nResultados da simulação:")
print("a) P(X = 3) =", prob_a)
print("b) P(X <= 2) =", prob_b)
print("c) P(X > 5) =", prob_c)
print("d) Média =", media)
print("   Variância =", variancia)
```
