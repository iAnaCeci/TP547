```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# parâmetros
# =========================
p = 0.3        # probabilidade de sucesso
n = 10000      # número de amostras

# =========================
# LCG (gerador uniforme)
# =========================
x = 1
a = pow(7, 5)
m = pow(2, 31) - 1

# =========================
# vetor de amostras
# =========================
geom = np.zeros(n)

# =========================
# simulação
# =========================
for i in range(n):
    falhas = 0

    while True:
        # gera uniforme (0,1)
        x = (a * x) % m
        u = x / m

        # teste de Bernoulli
        if u < p:
            break
        else:
            falhas += 1

    geom[i] = falhas

# =========================
# gráfico
# =========================
intervalo = range(0, 20)

plt.hist(geom, bins=intervalo, density=True, align='left', alpha=0.7, label="Simulação")

# eixo x
k = np.arange(0, 20)

# distribuição teórica
Fx = (1 - p)**k * p

# linha teórica
plt.plot(k, Fx, 'ro-', label="Teórico")

# =========================
# métricas
# =========================
print("Média simulada:", np.mean(geom))
print("Variância simulada:", np.var(geom))

# teórico
media_teorica = (1 - p) / p
variancia_teorica = (1 - p) / (p**2)

print("Média teórica:", media_teorica)
print("Variância teórica:", variancia_teorica)

# =========================
# labels
# =========================
plt.title("Distribuição Geométrica")
plt.xlabel("Número de falhas antes do sucesso")
plt.ylabel("Probabilidade")
plt.legend()
plt.grid()

plt.show()
```
